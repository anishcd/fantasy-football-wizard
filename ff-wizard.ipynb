```python
import os
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.width', None)
pd.options.display.max_rows = 4000

# Your existing code to load data_frames
years = range(2017, 2022)
folder_path = '/Users/anishdevineni/Desktop/projects/fantasy-football-wizard/data/fantasy-csv-data-master/yearly'
# Path to the folder containing injury data
injury_folder_path = '/Users/anishdevineni/Desktop/projects/fantasy-football-wizard/data/injury-data/NFL-Injury-Data-PFR--master/Data/'
data_frames = {}

for year in years:
    file_path = os.path.join(folder_path, f"{year}.csv")
    data_frames[year] = pd.read_csv(file_path)
    for name in data_frames[year]['Player']:
        if (name[-1] == '*' or name[-1] == '+') :
            name = name.replace('*', '')
            name = name.replace('+', '')

# print(data_frames[2021])
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.width', None)


def load_injury_data(player_name, year, injury_folder_path):
    injury_file_path = os.path.join(injury_folder_path, f"PFR_{year}_Injuries.csv")
    injury_data = pd.read_csv(injury_file_path, encoding='latin1')

    player_injury_data = injury_data[injury_data['Name'] == player_name]
    if not player_injury_data.empty:
        return player_injury_data
    else:
        return None


def get_player_stats_by_name(player_name, data_frames):
    player_stats_list = []

    for year, df in data_frames.items():
        player_data = df[df['Player'] == player_name]

        if not player_data.empty:
            player_stats = player_data.iloc[0].to_dict()
            player_stats['Year'] = year

            if year >= 2009 and year <= 2020:
                player_injury_data = load_injury_data(player_name, year, injury_folder_path)
                if player_injury_data is not None:
                    player_stats['Injury Data'] = player_injury_data.to_dict(orient='records')
                else:
                    player_stats['Injury Data'] = None

            player_stats_list.append(player_stats)

    if player_stats_list:
        return pd.DataFrame(player_stats_list)

    return pd.DataFrame()

# Usage example:
target_player_name = "Ameer Abdullah"  # Replace with the player's name you want to find

player_stats_df = get_player_stats_by_name(target_player_name, data_frames)

if not player_stats_df.empty:
    with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
        display(player_stats_df)
else:
    print("Player not found in the dataset.")
```

```python
import nfl_data_py as nfl
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

years = [2022]

df = nfl.import_seasonal_data(years)

# Function to deduplicate full string repeats
def deduplicate_string(s):
    if not s:
        return s
    length = len(s)
    for size in range(1, length // 2 + 1):  # size: size of the substring to test
        if length % size != 0:
            continue
        substring = s[:size]
        if substring * (length // size) == s:
            return substring
    return s  # Return original string if no repeating pattern is found

# Apply the function to the 'position' column
df['position'] = df['position'].apply(deduplicate_string)

# If needed, apply to other columns as well
df['player_name'] = df['player_name'].apply(deduplicate_string)
df['player_display_name'] = df['player_display_name'].apply(deduplicate_string)
df['position_group'] = df['position_group'].apply(deduplicate_string)

# Display the modified DataFrame
display(df)




```
